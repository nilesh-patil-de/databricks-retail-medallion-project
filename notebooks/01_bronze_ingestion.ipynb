```python
# ==========================================
# Project: Retail Medallion Architecture
# Layer: Bronze
# Purpose: Ingest raw CSV data into a Delta Bronze table
# ==========================================

from pyspark.sql import SparkSession

# ------------------------------------------
# Configuration
# ------------------------------------------

SOURCE_FILE = "/Volumes/retail_project/bronze/raw_files/Online%20Retail.csv"
BRONZE_TABLE = "retail_project.bronze.online_retail"

# ------------------------------------------
# Read Source File
# ------------------------------------------

df = (
    spark.read
    .option("header", "true")
    .option("inferSchema", "true")
    .csv(SOURCE_FILE)
)

# ------------------------------------------
# Data Validation
# ------------------------------------------

print("Source Row Count:", df.count())

df.printSchema()

# ------------------------------------------
# Write Bronze Table
# ------------------------------------------

(
    df.write
    .format("delta")
    .mode("overwrite")
    .saveAsTable(BRONZE_TABLE)
)

# ------------------------------------------
# Post-Load Validation
# ------------------------------------------

spark.sql(f"""
SELECT COUNT(*) AS row_count
FROM {BRONZE_TABLE}
""").show()

print("Bronze ingestion completed successfully.")
```
